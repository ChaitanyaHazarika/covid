```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
#Task 1 
#Sort the data as per total number of cases
sorteddf= df.sort_values(by= ["total_cases"])
sorteddf
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
lc= sorteddf["total_cases"].head(5)

least_cases= sorteddf["country"].head(5)


plt.xlabel("Country")
plt.xticks(rotation= 'vertical')
plt.ylabel("Cases")


label= least_cases
value= lc

plt.bar(label, value,width= 0.3,  color=("Blue", "Green", "Black", "Pink", "Red"))
```

```python
#Task 3
#Sort the data as per total number of deaths

totaldeaths= df.sort_values(by= ['total_deaths'])
totaldeaths
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph

hd= totaldeaths["total_deaths"].tail(5)
countryd= totaldeaths["country"].tail(5)

plt.xlabel("Countries")
plt.xticks(rotation= 'vertical')

plt.ylabel('Deaths')


label= countryd
value= hd

plt.bar(label, value, width=0.3, color=("Blue", "Green", "Black", "Pink", "Red"))
```

```python
#Task 5
#Sort the data as per active cases

ac= df.sort_values(by= ["active_cases"])
ac
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
achn= ac['active_cases'].tail(5)
countriessss= ac['country'].tail(5)

plt.xlabel('Countries')
plt.xticks(rotation= 'vertical')
plt.ylabel('Active cases bad bad')

label= countriessss
value= achn

plt.bar(label, value, width=0.25, color=("Blue", "Green", "Black", "Pink", "Red"))
```
